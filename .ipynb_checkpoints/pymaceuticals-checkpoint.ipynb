### Analysis

For this study we have received mouse metadata and observations for tumor volume under different drug regimens. The input data has been cleaned and merged. from the complete data set statistical summary has been generated. This analysis focuses on 4 drug regimen in particular, Capomulin, Ramicane, Infubinol, and Ceftamin.

From the bar plot showing total number observations for each drug regimen, we can conclude that drugs Capomulin and Ramicane have more observations compared to other drug. drug Propriva has least number of observations. Compared to Propriva other drugs have 20% more observations and Capomulin and Ramicane have about 33% more observations.

The sex distribution of the mice under observation is very balanced. There are 50.4% male mice and 49.6% female mice. This chart includes the mice data from cleaned data set.

There are 4 drug regimens which have been studied in depth in box plots. 
Ramicane: From the box plot we can see that there are more data points in 3rd and 4th quartile. The data is skewed towards top quartiles. The box plot is as per the statistical summary. Ramicane has low variance and standard deviation. Hence the smaller box and whiskers.

Capomulin: The data is very concentrated in 3rd quartile. The size of the box and whiskers are according to statistical summary, low variance and standard deviation.

Infubinol: The data points for Infubinol are more equally distributed compared to previous drugs. But the range of the tumor volumes is much higher compared to previous drugs. This has one outlier in first quartile.

Ceftamin: The data points are concentrated in 3rd and 4th quartile. This plot has shortest whiskers. This means most data points lie within 2nd and 3rd quartile. The range of tumor volume is much higher than Ramicane and Capomulin. But comparable with Infubinol.

From this box plots we can conclude that Ramicane and Capomulin are used for smaller tumor volumes whereas Infubinol and Ceftamin are used for higher tumor volumes.

Capomulin treatment of mouse s185
In this line chart we see very sharp decline in tumor volume for first time point at 10. Between 10 and 30 the gradient of the slope is higher. But again after 30 the tumor volume is reduced very rapidly.

Correlation between weight and tumor volume
We notice positive correlation between weight and tumor volume. Higher the weight of the mouse bigger tumor volume.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
print(f"No of records in mouse metadata: {len(mouse_metadata)}")
print(f"No of obervations in study results: {len(study_results)}")
study_data_complete.head()
```

```python
# Checking the number of mice.
noOfMice = study_data_complete['Mouse ID'].nunique()
noOfMice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice_df = pd.DataFrame(study_data_complete.groupby(['Mouse ID', 'Timepoint']).filter(lambda x: len(x) > 1))
mouseID_list = duplicate_mice_df['Mouse ID'].unique().tolist()

unique_study_data_complete = pd.DataFrame()
for mouseID in mouseID_list:
    unique_study_data_complete = study_data_complete.drop(study_data_complete[study_data_complete['Mouse ID'] == mouseID].index)


#unique_study_data_complete =unique_study_data_complete.rename(columns={"Tumor Volume (mm3)": "Tumor Volume"})
unique_study_data_complete
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean = unique_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = unique_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
var = unique_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std = unique_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = unique_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

study_result_summary_1 = pd.DataFrame(
       {"Mean": mean,
        "Median": median,
        "var": var,
        "std": std,
        "SEM": sem
       })
study_result_summary_1
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
study_result_summary_2 = pd.DataFrame(unique_study_data_complete.groupby(['Drug Regimen']).agg({'Tumor Volume (mm3)':['mean', 'median', 'var', 'std', 'sem']}))
study_result_summary_2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
combined_data = pd.DataFrame(unique_study_data_complete.loc[(unique_study_data_complete['Drug Regimen'] == 'Capomulin') |
                                                             (unique_study_data_complete['Drug Regimen'] == 'Ramicane') |
                                                             (unique_study_data_complete['Drug Regimen'] == 'Infubinol') |
                                                             (unique_study_data_complete['Drug Regimen'] == 'Ceftamin')])

# Start by getting the last (greatest) timepoint for each mouse
g = pd.DataFrame(combined_data.groupby('Mouse ID')['Timepoint'].max())
g.reset_index(inplace = True)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor = pd.merge(g, combined_data, on =['Mouse ID', 'Timepoint'], how ='inner')
final_tumor
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment = combined_data['Drug Regimen'].unique()
data = pd.DataFrame()

for drug_name in treatment:
    treatment_data = final_tumor.loc[final_tumor['Drug Regimen'] == drug_name]
    data[drug_name] = treatment_data['Tumor Volume (mm3)'].to_list()
    tumor_vol = treatment_data['Tumor Volume (mm3)']
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    outlier = []
    for x in tumor_vol:
        if (x < lower_bound) | (x > upper_bound):
            outlier.append(x)
        
    print(f"{drug_name} outliers: {outlier}")    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
data.boxplot().set_ylabel('Tumor Volume (mm3)')
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
s185_data = pd.DataFrame(unique_study_data_complete.loc[(unique_study_data_complete['Drug Regimen'] == 'Capomulin') &
                                                             (unique_study_data_complete['Mouse ID'] == 's185')])
s185_data

s185_line, = plt.plot(s185_data['Timepoint'], s185_data['Tumor Volume (mm3)'], label="Capomulin")
plt.title("Capomulin treatment of mouse s185")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

Capomulin_data = pd.DataFrame(unique_study_data_complete.loc[unique_study_data_complete['Drug Regimen'] == 'Capomulin'])
g = pd.DataFrame(Capomulin_data.groupby(['Mouse ID', 'Weight (g)'])['Tumor Volume (mm3)'].mean())
g.reset_index(inplace = True)
#g

plt.scatter(g['Weight (g)'], g['Tumor Volume (mm3)'], marker="o")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(g['Weight (g)'],g['Tumor Volume (mm3)'])
print(f"The correlation between weight(g) and tumor volume(mm3) is {round(correlation[0],2)}")

x_values = g['Weight (g)']
y_values = g['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```
